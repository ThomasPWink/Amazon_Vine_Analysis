```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("M17-Amazon-Challenge").config("spark.driver.extraClassPath","/content/postgresql-42.2.16.jar").getOrCreate()
```

```python
from pyspark import SparkFiles
url = "https://s3.amazonaws.com/amazon-reviews-pds/tsv/amazon_reviews_us_Sports_v1_00.tsv.gz"
spark.sparkContext.addFile(url)
df = spark.read.option("encoding", "UTF-8").csv(SparkFiles.get(""), sep="\t", header=True, inferSchema=True)
df.show()
```

```python
# Create the vine_table. DataFrame
from pyspark.sql.types import IntegerType
vine_df = df.select(["review_id", "star_rating", "helpful_votes", "total_votes", "vine", "verified_purchase"]).withColumn("star_rating", df["star_rating"].cast(IntegerType()))
vine_df.show()
```

```python
vine_20_df = vine_df.filter("total_votes >=20")
vine_20_df.show()
```

```python
helpful_vine_df = vine_20_df.filter(vine_20_df["helpful_votes"]/vine_20_df["total_votes"] >= .5)
helpful_vine_df.show()
```

```python
vine_paid_df = helpful_vine_df.filter(helpful_vine_df["vine"] == "Y")
vine_paid_df.show()
```

```python
vine_unpaid_df = helpful_vine_df.filter(helpful_vine_df["vine"] == "N")
vine_unpaid_df.show()
```

```python
# total 5 star rating for paid
vine_paid_df.filter(vine_paid_df["star_rating"] == "5").count()
```

```python
# total reviews for paid
vine_paid_df.count()
```

```python
# Percentage of 5 star reviews paid
vine_paid_df.filter(vine_paid_df["star_rating"] == "5").count()/vine_paid_df.count()
```

```python
# 5 star reviews for unpaid
vine_unpaid_df.filter(vine_unpaid_df["star_rating"] == "5").count()
```

```python
#total reviews for unpaid
vine_unpaid_df.count()
```

```python
# Percentage of 5 star reviews unpaid
vine_unpaid_df.filter(vine_unpaid_df["star_rating"] == "5").count()/vine_unpaid_df.count()
```
